```python
!pip install -q -U google-genai

from google.genai import Client
from google.colab import userdata

api_key = userdata.get('GOOGLE_API_KEY')
client = Client(api_key=api_key)

book_list = ["The Great Gatsby", "A Brief History of Time", "Neuromancer", "The 7 Habits of Highly Effective People"]

prompt = f"Categorize these books by genre and give a 1-sentence summary for each: {book_list}"

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=prompt
)

print(response.text)
print(f"Tokens Used: {response.usage_metadata.total_token_count}")
```

I successfully used the google-genai library to interact with the Gemini-3-Flash model via Python. I processed a batch of four different books in a single API call to demonstrate efficiency. I monitored the total_token_count (627 tokens). In a production environment, tracking this metadata is essential for cost management and monitoring rate limits.
